```python
# Must keep header none to drop breaklines
tweets_data = pd.read_csv("datasources/data.csv", header=None, skiprows=[0])
tweets_data
```

```python
# Convert date column from object to datetime format
from datetime import datetime
tweets_data[0] = pd.to_datetime(tweets_data[0], format='%Y-%m-%d %H:%M:%S')
tweets_data.dtypes
```

```python
# Add a weekday column
tweets_data[3] = tweets_data[0].dt.day_name()

# Add an hour column
tweets_data[4] = tweets_data[0].dt.strftime('%H')

# Convert date format
tweets_data[0] = tweets_data[0].dt.strftime('%Y-%m-%d')


# Change column names and reorder columns
tweets_data.columns = ['created_date','id','tweet','weekday','hours']
tweets_data = tweets_data [['created_date','weekday','hours','id','tweet']]
```
