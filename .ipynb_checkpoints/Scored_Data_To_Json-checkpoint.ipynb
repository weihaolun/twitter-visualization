```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib
import json
```

```python
# load cleaned scored tweets data
scored_tweets = pd.read_json("datasources/cleaned_scored_tweets.json")
scored_tweets
```

```python
scored_tweets.dtypes
```

```python
# Random select 200 samples per hour
hourly_sample = scored_tweets.groupby(["created_date","hours"], sort=False).sample(n=200, random_state=1)
hourly_sample
```

```python
# Save to new json
hourly_sample.to_json("datasources/all_weeks_data.json", orient="records")
```

```python
# Load tweet count data
tweets_count = pd.read_json("datasources/tweets_count.json")
tweets_count
```

```python
# check datatypes
tweets_count.dtypes
```

```python
# Assign week number to data
conditions = [
    (tweets_count["created_date"] <= "2021-11-07"),
    (tweets_count["created_date"] <= "2021-11-14")]

values = ["WEEK01", "WEEK02"]

tweets_count["weeks"] = np.select(conditions, values)

tweets_count
```

```python
# Save to json
tweets_count.to_json("datasources/weekly_tweets_counts.json", orient="records")
```
