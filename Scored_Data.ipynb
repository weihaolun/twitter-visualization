```python
import pandas as pd
from datetime import datetime
import matplotlib
import json
```

```python
# Must keep header none to drop breaklines
scored_tweets = pd.read_csv("datasources/scored_tweets.csv", header=None, skiprows= [0])
scored_tweets
```

```python
scored_tweets.dtypes
```

```python
# Change column names and reorder columns
scored_tweets.columns = ['created_date','weekday','hours','id','tweet', 'score']
scored_tweets
```

```python
scored_tweets.to_csv("datasources/scored_data.csv", index=True)
```

```python
scored_tweets.to_json("datasources/scored_data.json", orient="records")
```

```python
hourly_sample = scored_tweets.groupby(["weekday","hours"]).sample(n=200, random_state=1)
```

```python
hourly_sample.to_csv("datasources/hourly_sample.csv", index=True)
```

```python
hourly_sample.to_json("datasources/hourly_sample.json", orient="records")
```
